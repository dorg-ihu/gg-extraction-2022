```python
def print_results(result):
  for i in range(len(result['answers'])):
    print(f'Answer:\n{result["answers"][i].answer}')
    print(f'Offsets in document {result["answers"][i].offsets_in_document}')
    print(f'Offsets in context {result["answers"][i].offsets_in_context}')
    print(f'Context:\n{result["answers"][i].context}')
    print(f'Score ==> {result["answers"][i].score}')
    print("\n")
    print("##"*10)


def get_results(result, entity):
  answers = []
  for i in range(len(result['answers'])):
    print(fuzz.partial_ratio(entity, result['answers'][i].context))
    if fuzz.partial_ratio(entity, result['answers'][i].context) > 85:
      answers.append(result['answers'][i])
  
  if answers:
    max_score = 0
    for i in range(len(answers)):
      if answers[i].score > max_score:
        answer = answers[i].answer
        max_score = answers[i].score
    return answer
  else:
    return ""


def get_query(entity):
  base_query = "Τι αρμοδιότητες έχει "
  if any(x in entity for x in ["Τμήμα", "ΤΜΗΜΑ"]):
    query = base_query + 'το ' + entity + ';'
  elif any(x in entity for x in ["Διεύθυνση", "ΔΙΕΥΘΥΝΣΗ", "Γραμματεία", "ΓΡΑΜΜΑΤΕΙΑ"]):
    query = base_query + 'η ' + entity + ';'
  else:
    query = base_query + entity + ';'
  return query

```

```python
import re

txt = "(α) Η παροχή νομικής υποστήριξης και οδηγιών νομικής φύσεως"

def get_lvl_pattern(txt):
  pattern = re.findall(r"^[ \(]*[α-ωΑ-Ω0-9]+[\.\)] ", txt)

  if pattern:
    return re.sub(r'[^Α-ΩA-Zα-ωa-z0-9]+', '', pattern[0])
  else:
    return ""

lvl = get_lvl_pattern(txt)
print(lvl)
```
